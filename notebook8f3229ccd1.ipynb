**1.library import**

```python
# ðŸ“¥ Import Libraries

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

import warnings
warnings.filterwarnings("ignore")
```

**2.Data loading**

```python
train_path = "/kaggle/input/competitions/house-prices-advanced-regression-techniques/train.csv"
test_path = "/kaggle/input/competitions/house-prices-advanced-regression-techniques/test.csv"

train = pd.read_csv(train_path)
test = pd.read_csv(test_path)

print("Train shape:", train.shape)
print("Test shape:", test.shape)

train.head()
```

**#Data Exploration**

```python
train.info()
```

```python
train.describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(train["SalePrice"], kde=True)

plt.title("Sale Price Distribution")
plt.show()
```

**4.data preprocessing**

```python
# Save Id
train_ID = train["Id"]
test_ID = test["Id"]

# Separate target
y = train["SalePrice"]

train = train.drop(["Id", "SalePrice"], axis=1)
test = test.drop(["Id"], axis=1)
```

```python
# Combine datasets
full = pd.concat([train, test])
```

```python
# Fill missing numeric values
num_cols = full.select_dtypes(include=np.number).columns
full[num_cols] = full[num_cols].fillna(full[num_cols].median())

# Fill missing categorical values
cat_cols = full.select_dtypes(include="object").columns
full[cat_cols] = full[cat_cols].fillna(full[cat_cols].mode().iloc[0])
```

```python
# Convert categorical to numeric
full = pd.get_dummies(full)
```

```python
# Split back
X = full.iloc[:len(y)]
X_test = full.iloc[len(y):]

print("Training features:", X.shape)
print("Test features:", X_test.shape)
```

**Model Training**

Correlation heatmap

```python
model = RandomForestRegressor(
    n_estimators=300,
    random_state=42,
    n_jobs=-1
)

model.fit(X, y)

print("Model training completed")
```

Prediction


```python
predictions = model.predict(X_test)

print(predictions[:5])
```

Output File


```python
output = pd.DataFrame({
    "Id": test_ID,
    "SalePrice": predictions
})

output.to_csv("submission.csv", index=False)

output.head()
```
